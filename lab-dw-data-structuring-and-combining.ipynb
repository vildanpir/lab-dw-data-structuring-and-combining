```python
import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
print(df1.head())
print(df2.head())
print(df3.head())
```

```python
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.replace(" ", "_")
df3.columns = df3.columns.str.lower().str.replace(" ", "_")
```

```python
df1 = df1.rename(columns={"st": "state"})
df2 = df2.rename(columns={"st": "state"})
# df3 zaten state, ona dokunmaya gerek yok
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df3.isnull().sum()
```

```python
df = pd.concat([df1, df2, df3])
df = df.reset_index(drop=True)
df.shape
```

```python
df.columns
```

```python
df["gender"] = df["gender"].replace({"Femal": "F", "female": "F", "Male": "M"})
```

```python
df["education"] = df["education"].replace({"Bachelors": "Bachelor"})
```

```python
def fix_clv(value):
    if isinstance(value, str):
        return value.replace("%", "")
    else:                               
        return value              
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].apply(fix_clv) 
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].astype(float)
```

```python
df["vehicle_class"] = df["vehicle_class"].replace({"Sports Car": "Luxury", "Luxury SUV": "Luxury", "Luxury Car": "Luxury"})
```

```python
def fix_complaints(value):
    if "/" in str(value):
        return str(value).split("/")[1]
    else:
        return value
```

```python
df["number_of_open_complaints"] = df["number_of_open_complaints"].apply(fix_complaints)
```

```python
df["number_of_open_complaints"] = df["number_of_open_complaints"].astype(float)
```

```python
df = df.dropna(how="all")
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```
